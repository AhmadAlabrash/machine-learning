```python
#to make horizntal join to two table
df1 = pd.DataFrame({
    'city':['delhi','newdelhi','mombi'],
    'tempreture':[45,33,24]
})
df2 = pd.DataFrame({
    'city':['delhi','newdelhi','mombi'],
    'wind':[444,120,224]
})
pd.concat([df1,df2],axis=1)
```

```python
#to make vertical join to two table
df1 = pd.DataFrame({
    'city':['delhi','newdelhi','mombi'],
    'tempreture':[45,33,24]
})
df2 = pd.DataFrame({
    'city':['delhi','newdelhi','mombi'],
    'wind':[444,120,224]
},index=[2,1,0])
pd.concat([df1,df2])
```

```python
#to make merge (inner join) to two table on forginkey  (inner join & outer join)
df1 = pd.DataFrame({
    'city':['delhi','newdelhi','mombi'],
    'tempreture':[45,33,24]
})
df2 = pd.DataFrame({
    'city':['delhi','newdelhi','mombi'],
    'wind':[444,120,224]
})
pd.merge(df1,df2,on='city')
```

```python
#to make merge to two table on forginkey (inner join & outer join)
df1 = pd.DataFrame({
    'city':['delhi','newdelhi','mombi'],
    'tempreture':[45,33,24]
})
df2 = pd.DataFrame({
    'city':['delhi','newdelhi','mombi'],
    'wind':[444,120,224]
})
pd.merge(df1,df2,on='city',how='right') #you also have outer and right join and left join
```
