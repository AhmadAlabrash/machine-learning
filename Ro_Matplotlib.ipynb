```python
xpos = ['Googl','Amzon','Tesla','meta']
xCover = np.arange(len(xpos))

revenue = [200,119,170,130]
profit = [10,19,14,18]

plt.title('USA Companies 2022')

plt.xticks(xCover,xpos)
plt.bar( xCover +0.2, revenue , width=0.4, color='#FD841F',label = 'Revenue')
plt.bar( xCover -0.2, profit , width=0.4 ,label = 'Profit')
plt.legend(shadow=True)

plt.savefig('USA companies22')
```

```python
plt.yticks(xCover,xpos , )
plt.barh(xCover+0.2  , revenue,label = 'Revenue')
plt.barh(xCover-0.2  , profit, label = 'Profit')
plt.legend(shadow=True)
```

```python
blooodMen=[77,88,85,83,89,80,90]
blooodWomen=[81,68,75,80,79,70,70]

plt.hist([blooodMen,blooodWomen],bins=[67,70,73,76,79,82,85,88,91,],rwidth=1.995 ,color=['blue','red'],label=['Men','Women'] , orientation='horizontal')
plt.legend(shadow=True)
```

```python
cash=[2000,1200,500,300,700]
cash_y =['Home','Food','Car','Phone','Others']
plt.pie(cash, labels= cash_y, radius = 2.5 ,autopct='%0.1f%%' ,explode=[0,0.2,0,0.1,0.6] ,startangle=20)

plt.savefig('personal expenses',bbox_inches='tight',pad_inches=2 ,transparent=True)
```
