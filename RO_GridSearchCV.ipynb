```python
import pandas as pd
from sklearn.model_selection import GridSearchCV    
from sklearn.linear_model import LinearRegression 
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

from sklearn.naive_bayes import MultinomialNB
```

```python
df = pd.read_csv('sp.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer
le =LabelEncoder()
df['spam'] = le.fit_transform(df.Category )
co =CountVectorizer()                                                                                        
y=df['spam']
x_co = co.fit_transform(df.Message)
x_co.toarray()
```

```python
clf =GridSearchCV(SVC(gamma='auto'),{
    'kernel':['rbf','liner'],
    'C':[1,10,20,30]
},return_train_score=False
)              
clf.fit(x_co,y)
```

```python
clf.cv_results_
```

```python
pd.DataFrame(clf.cv_results_)
```

```python
#we will ccccccchange dataset here
df=pd.read_csv('/content/titanic.csv')
df =df.dropna()
```

```python
df_para ={
   'svm' :{ 'model': SVC(),
    'params':{
        'kernel':['rbf','liner'],
        'C':[1,20,30]
    }},
     'ra' :{ 'model': RandomForestClassifier(),
    'params':{
        'n_estimators':[5,10,15,20,25,30],
        'max_depth':[10,20,30]
    }}
    ,
     'nb' :{ 'model': MultinomialNB(),
    'params':{
        
        'alpha':[0.001,0.01,30]
    }}
}
```

```python
res=[]
for useless ,mo in df_para.items() :
                    
    clf =GridSearchCV(mo['model'],mo['params'],cv=5,return_train_score=False)              
    clf.fit(df[['Parch','Pclass','Age']],df.Survived)
    res.append ({'bestParams':clf.best_params_ ,'bestscore':clf.best_score_})
   
```

```python
pd.DataFrame(res)
```
